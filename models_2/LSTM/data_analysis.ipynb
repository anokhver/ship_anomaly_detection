```python
import pandas as pd
```

```python
df = pd.read_parquet('data/LSTM_preprocessed.parquet')
```

```python
df_KIEL = df[df['start_port'] == 'KIEL'].copy()
df_BREMERHAVEN = df[df['start_port'] == 'BREMERHAVEN'].copy()
```

```python
average_time = df_KIEL['end_time'] - df_KIEL['start_time']
average_time_mean = average_time.mean()
average_time_mean
```

```python
df_KIEL['time_diff'] = df_KIEL.groupby('trip_id')['time_stamp'].diff()
average_interval_per_trip = df_KIEL.groupby('trip_id')['time_diff'].mean()
overall_average_interval = average_interval_per_trip.mean()
overall_average_interval
```

```python
average_time = df_BREMERHAVEN['end_time'] - df_BREMERHAVEN['start_time']
average_time_mean = average_time.mean()
average_time_mean
```

```python
df_BREMERHAVEN['time_diff'] = df_BREMERHAVEN.groupby('trip_id')['time_stamp'].diff()
average_interval_per_trip = df_BREMERHAVEN.groupby('trip_id')['time_diff'].mean()
overall_average_interval = average_interval_per_trip.mean()
overall_average_interval
```
