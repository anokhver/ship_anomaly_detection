```python
import pandas as pd
import os
```

```python
file_path = '../../data/fixed_notnum.parquet'
output_file_path = '../../data/fixed_num.parquet'

if not os.path.exists(file_path):
    print(f"File not found: {file_path}")

df = pd.read_parquet(file_path)
df.head()
```

1. Length
2. Breadth
3. Draught

```python
print((df['Length'] == 0).sum())
print((df['Breadth'] == 0).sum())
print(((df['Length'] == 0) & (df['Breadth'] == 0)).sum())
print((df['Draught'] == 0).sum())
df[(df['Length'] == 0) & (df['Breadth'] == 0)]['shiptype'].value_counts()
```

```python
print((df['shiptype'] == 0).sum() - ((df['Length'] == 0) & (df['Breadth'] == 0) & (df['shiptype'] == 0 )).sum())
print((df['shiptype'] == 70).sum() - ((df['Length'] == 0) & (df['Breadth'] == 0) & (df['shiptype'] == 70 )).sum())
print((df['shiptype'] == 71).sum() - ((df['Length'] == 0) & (df['Breadth'] == 0) & (df['shiptype'] == 71 )).sum())
```

So we can use mean

```python
import numpy as np
df['Length'] = df['Length'].replace(0, np.nan)
df['Breadth'] = df['Breadth'].replace(0, np.nan)
df['Draught'] = df['Draught'].replace(0, np.nan)

# Step 2: Calculate mean by shiptype (automatically ignores NaN)
length_means = df.groupby('shiptype')['Length'].transform('mean')
breadth_means = df.groupby('shiptype')['Breadth'].transform('mean')
draught_means = df.groupby('shiptype')['Draught'].transform('mean')

# Step 3: Fill NaN with shiptype means
df['Length'] = df['Length'].fillna(length_means)
df['Breadth'] = df['Breadth'].fillna(breadth_means)
df['Draught'] = df['Draught'].fillna(draught_means)

print((df['Length'] == 0).sum())
print((df['Breadth'] == 0).sum())
print(((df['Length'] == 0) & (df['Breadth'] == 0)).sum())
print((df['Draught'] == 0).sum())
```

```python
df = df.drop_duplicates()
```
